# Plotting Activation Functions

```python
x = np.array([i for i in range(-11,11)])
```

### Sigmoid Function:

$$ 
\sigma (x) = \frac{1}{1+e^{-x}}
$$

```python
y = sigmoid(x)
plt.plot(x,y)
plt.xlabel("x")
plt.ylabel("sigmoid(x)")
plt.grid()
```

### Hyperbolic Tangent Function:

$$
tanh(x) = \frac{e^{x} - e^{-x}}{e^{x}+e^{-x}} 
$$

```python
y = tanh(x)
plt.plot(x,y)
plt.xlabel("x")
plt.ylabel("tanh(x)")
plt.grid()
```

### Rectified Linear Unit:

$$
ReLU(x) = max(0, x)
$$

```python
y = np.array([relu(i) for i in x])
plt.plot(x,y)
plt.xlabel("x")
plt.ylabel("ReLU(x)")
plt.grid()
```
